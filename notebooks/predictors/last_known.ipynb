```python
import pandas as pd
from sklearn.metrics import mean_squared_error, mean_absolute_error, root_mean_squared_error
```

```python
df = pd.read_csv('../data/SBER_250914_250915.csv', sep=';')

df["date_str"]=df["<DATE>"].apply(lambda x: str(x).zfill(6)).astype(str)
df["time_str"]=df["<TIME>"].apply(lambda x: str(x).zfill(6)).astype(str)
df["datetime_str"]=df["date_str"]+df["time_str"]
df["time_utc"]= pd.to_datetime(df["datetime_str"], format='%y%m%d%H%M%S', utc=True)
df = df.set_index("time_utc")
df.drop(["date_str", "time_str", "datetime_str", "<DATE>", "<TIME>", "<OPEN>", "<HIGH>", "<LOW>"], axis=1, inplace=True)
df.rename(columns={"<CLOSE>": "close", "<VOL>": "volume"}, inplace=True)

#df = df.tail(200)
df["close_predict"] = df["close"].shift()
df.dropna(inplace=True )
```

```python
mse = mean_squared_error(df["close"], df["close_predict"])
mae = mean_absolute_error(df["close"], df["close_predict"])
rmse = root_mean_squared_error(df["close"], df["close_predict"])

print(f"MSE={mse} RMSE={rmse} MAE={mae}")
```
