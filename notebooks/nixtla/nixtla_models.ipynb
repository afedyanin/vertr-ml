```python
from statsforecast import StatsForecast
from app.models.nixtla_models import SeriesItem, series_to_df, df_to_forecast_items, keys_to_stats_models
from datetime import datetime
```

```python
series = [
    SeriesItem(unique_id='SBER', ds=datetime(2020, 11, 1, 14, 45, 00), y=100),
    SeriesItem(unique_id='SBER', ds=datetime(2020, 11, 1, 14, 46, 00), y=120),
    SeriesItem(unique_id='SBERP', ds=datetime(2020, 11, 1, 14, 50, 00), y=160),
    SeriesItem(unique_id='SBERP', ds=datetime(2020, 11, 1, 14, 51, 00), y=170),
    SeriesItem(unique_id='SBER', ds=datetime(2020, 11, 1, 14, 47, 00), y=130),
    SeriesItem(unique_id='SBER', ds=datetime(2020, 11, 1, 14, 48, 00), y=140),
    SeriesItem(unique_id='SBER', ds=datetime(2020, 11, 1, 14, 49, 00), y=150),
    SeriesItem(unique_id='SBER', ds=datetime(2020, 11, 1, 14, 50, 00), y=160),
    SeriesItem(unique_id='SBER', ds=datetime(2020, 11, 1, 14, 51, 00), y=170),
    SeriesItem(unique_id='SBER', ds=datetime(2020, 11, 1, 14, 52, 00), y=180),
    SeriesItem(unique_id='SBERP', ds=datetime(2020, 11, 1, 14, 49, 00), y=150),
    SeriesItem(unique_id='SBERP', ds=datetime(2020, 11, 1, 14, 52, 00), y=180),
]

df = series_to_df(series)

#df
```

```python
model_keys = [
    'naive',
    'auto_arima',
    #'auto_ets',
    #'auto_ces',
    'auto_theta',
    'random_walk',
    'history_average',
]

models = keys_to_stats_models(model_keys)
sf = StatsForecast(models=models, freq='min')
sf.fit(df=df)
forecast_df = sf.predict(h=1)
forecast_df
```

```python
items = df_to_forecast_items(forecast_df, model_keys)
items
```
